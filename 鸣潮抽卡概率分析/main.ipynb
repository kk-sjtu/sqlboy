```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# 设置Matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
def simulate_draw():
    results = []
    for i in range(1,11):
        if i == 10:
            results.append((i,"A卡（保底）"))
            break
        random_number = random.random()
        if random_number < 0.1:
            results.append((i,"A卡"))
            break
        else:
            results.append((i,"非中"))
    return results

num_simulations = 1000
all_result = []
for _ in range(num_simulations):
    all_result.extend(simulate_draw())
df = pd.DataFrame(all_result,columns=["次数","抽卡结果"])
plt.figure(figsize=(10,6))
plt.hist(df["次数"],bins=np.arange(1,12,1),align="left",rwidth=0.7)
plt.xlabel("抽卡次数")
plt.ylabel("出现次数")
plt.title("抽卡结果分布")
plt.xticks(range(1,11))
plt.grid(axis="y",linestyle="--",alpha=0.6)
plt.show()

df.to_csv("result.csv",index=False)
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# 设置Matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 模拟抽卡函数
def simulate_draw():
    for i in range(1, 11):
        if i == 10:
            return 10  # 第10次抽卡（保底）
        random_number = random.random()
        if random_number < 0.1:
            return i  # 抽中A卡的次数
    return 10  # 确保保底机制生效

# 进行多次模拟
num_simulations = 100000  # 模拟1000次
draw_counts = [simulate_draw() for _ in range(num_simulations)]

# 统计每个抽卡次数的出现次数
draw_summary = pd.Series(draw_counts).value_counts().sort_index()

# 可视化结果
plt.figure(figsize=(10, 6))
plt.bar(draw_summary.index, draw_summary.values, color="skyblue", edgecolor="black")
plt.title("不同抽卡次数下抽中A卡的次数分布")
plt.xlabel("抽卡次数")
plt.ylabel("抽中次数")
plt.xticks(range(1, 11))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 导出为CSV文件
draw_summary.to_csv("draw_summary.csv", header=["次数"], index_label="抽卡次数")
```

```python
# 概率0.1时的累计分布函数
import random
import pandas as pd
import matplotlib.pyplot as plt

# 设置Matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 模拟抽卡函数
def simulate_draw():
    for i in range(1, 11):
        if i == 10:
            return 10  # 第10次抽卡（保底）
        random_number = random.random()
        if random_number < 0.1:
            return i  # 抽中A卡的次数
    return 10  # 确保保底机制生效

# 进行多次模拟
num_simulations = 100000  # 模拟1000次
draw_counts = [simulate_draw() for _ in range(num_simulations)]

# 统计每个抽卡次数的出现次数
draw_summary = pd.Series(draw_counts).value_counts().sort_index()

# 计算累计分布函数
cdf = draw_summary.cumsum() / num_simulations

# 可视化结果
plt.figure(figsize=(10, 6))
plt.plot(cdf.index, cdf.values, marker="o", color="skyblue")
plt.title("抽中A卡的累计分布函数")
plt.xlabel("抽卡次数")
plt.ylabel("累计概率")
plt.xticks(range(1, 11))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 导出为CSV文件
cdf.to_csv("draw_cdf.csv", header=["累计概率"], index_label="抽卡次数")
```
